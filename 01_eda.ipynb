```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style="whitegrid")

df=pd.read_csv("data/titanic.csv")

plt.figure(figsize=(6,4))

sns.countplot(data=df, x="Survived", hue= "Survived", palette=["#e74c3c","#2ecc71"], legend= False )

plt.title("Overall Surviror Count")
plt.ylabel("Number of Passengers")
plt.xticks([0,1], ["Did not Survive", "Survived"])
plt.show()

print(f"Baseline Survival Rate: {df['Survived'].mean():.2%}")
```

```python
plt.figure(figsize=(6,4))

sns.barplot(data= df, x="Sex", y="Survived", hue= "Sex", palette=["#3498db","#e91e63"], legend= False )

plt.title("Survival Rate by Gender")
plt.ylabel("Probability of Survival")
plt.ylim(0,1)
plt.show()

print("Survival Rate by Gender")
print(df.groupby("Sex")["Survived"].mean().apply(lambda x: f"{x:.2%}"))
```

```python
plt.figure(figsize=(6,4))

sns.barplot(data= df, x="Pclass", y="Survived", hue= "Pclass", palette= "viridis", legend= False )

plt.title("Survival Rate by Pclass")
plt.ylabel("Probability of Survival")
plt.ylim(0,1)

plt.xticks([0,1,2], ["1st Class", "2nd Class", "3rd Class"])
plt.show()

print("Survival Rate by  Passenger Class")
print(df.groupby("Pclass")["Survived"].mean().apply(lambda x: f"{x:.2%}"))
```

```python
plt.figure(figsize=(8,5))

sns.barplot(data= df, x="Pclass", y="Survived", hue= "Sex", palette=["#3498db","#e91e63"])

plt.title("Survival Rate by Pasenger Class and Gender")
plt.ylabel("Probability of Survival")
plt.ylabel("Probability of Survival")
plt.ylim(0,1)
plt.xticks([0,1,2], ["1st Class", "2nd Class", "3rd Class"])

plt.legend(title="Gender", loc="upper left")

plt.show()

print("Survival Rate by Class and Gender")
print(df.groupby(["Pclass", "Sex"])["Survived"].mean().unstack())
```

```python
plt.figure(figsize=(10,5))

sns.histplot(data=df,x="Age", hue="Survived",multiple="stack",palette=["#e74c3c","#2ecc71"], bins=30,kde=True)

plt.title("Passenger Age Distribution by Survival")
plt.xlabel("Age(years)")
plt.ylabel("Number of Passengers")

plt.legend(title="Status", labels=["Survived", "Not Survived"])
plt.show()
```
